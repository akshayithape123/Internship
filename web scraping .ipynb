```python
names=[]
for i in soup.find_all('div',class_="presidentListing"):
     names.append(i.text.split('\n')[1])

term=[]
for i in soup.find_all('div',class_="presidentListing"):
     term.append(i.text.split('\n')[2])
```

```python
for i in range(len(names)):
    names[i] = names[i].split('(')[0].strip()
    


for i in range(len(term)):
    term[i]=term[i].split(':')[1]
    
names_of_presidents =pd.DataFrame({'President names':names,'Term of office':term})
names_of_presidents
```

# Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame

```python
head=requests.get('https://www.cnbc.com/world/?region=world')
head
```

```python
soup=BeautifulSoup(head.content,'html.parser')
soup
```

```python
articles = soup.find_all('div', class_='Card-titleContainer')
```

```python
headlines = []
times = []
links = []
for article in articles:
    headline = article.find('a').text.strip()
    time_element = article.find('time')
    if time_element is not None:
        time = time_element['datetime']
    else:
        time = None
    link = article.find('a')['href']
    headlines.append(headline)
    times.append(time)
    links.append(link)
    
data = {'Headline': headlines, 'Time': times, 'News Link': links}
df = pd.DataFrame(data)
df
```

# 9) Write a python program to scrape mentioned details from dineout.co.in and make data frame

```python
rest=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup=BeautifulSoup(rest.content)
```

```python
header=[]
for i in soup.find_all('a',class_='restnt-name ellipsis'):
    header.append(i.text)
    
dish=[]
for i in soup.find_all('span',class_='double-line-ellipsis'):
    dish.append(i.text.split('|')[1])
    
Location=[]
for i in soup.find_all('div',class_='restnt-loc ellipsis'):
    Location.append(i.text)
    
Ratings=[]

for i in soup.find_all('div',class_='restnt-rating rating-4'):
    Ratings.append(i.text)
    
url=[]
for i in soup.find_all('img',class_='no-img'):
    url.append(i.get('data-src'))
    
df=pd.DataFrame({'Restaurant name':header  ,'Cuisine': dish  ,'Location': Location  ,'Ratings': Ratings  ,'Image URL':url})
df
```
