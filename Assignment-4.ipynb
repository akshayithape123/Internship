# 1. Scrape the details of most viewed videos on YouTube from Wikipedia.       Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views


```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")     
from selenium.webdriver.common.by import By 
import time
from selenium.common.exceptions import  StaleElementReferenceException, NoSuchElementException
```

```python
driver=webdriver.Chrome(r'chromedriver')
driver
```

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
rank=[]
name=[]
artist=[]
upload=[]
views=[]
```

```python
try:
    rank_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
    for i in rank_tags:
        rank.append(i.text)
except NoSuchElementException:
    rank.append('No Deatails')
    
    
try:
    name_tags= driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
    for i in name_tags:
        name.append(i.text)
        
except NoSuchElementException:
    name.append('No Details')
    
try:
    artist_tags= driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
    for i in artist_tags:
        artist.append(i.text)
except NoSuchElementException:
    artist.append('No Details')
    
try:
    upload_tags= driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
    for i in upload_tags:
        upload.append(i.text)
        
except NoSuchElementException:
    upload.append('No Details')
try:
    views_tags= driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
    for i in views_tags:
        views.append(i.text)
        
except NoSuchElementException:
    views.append('No Details')
```

```python
print(len(rank),len(name),len(artist),len(upload),len(views))
```

```python
df=pd.DataFrame({'Rank':rank,'Name':name,'Artist':artist,'Upload':upload,'Views In Billions':views})
df
```

```python
driver.close()
```
